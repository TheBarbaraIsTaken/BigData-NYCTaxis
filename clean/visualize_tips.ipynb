```python
import pandas as pd
import matplotlib.pyplot as plt

def process_line(line):
    return [int(line[0]), int(line[1]), int(line[2]), int(line[3]), float(line[4]), float(line[5])]

# Read data
data = None
with open("tip_by_year.txt", "r") as f:
    columns = f.readline().split()
    data = [process_line(line.strip().split()) for line in f.readlines()]

df = pd.DataFrame(data, columns=columns)
df
```

```python
x_values = df.year
y_values = df.total

# Plot the values
plt.plot(x_values, y_values, marker='o', linestyle='-', color='b')
plt.title(r'The total number of trips in each year')
plt.xlabel('year')
plt.ylabel('#rides')
plt.grid(True, alpha=0.8)
plt.show()
```

```python
x_values = df.year
y_values = df.non_zero_tip_count

# Plot the values
plt.plot(x_values, y_values, marker='o', linestyle='-', color='b')
plt.title('The number of trips where tip was given')
plt.xlabel('year')
plt.ylabel('#rides with tip')
plt.grid(True, alpha=0.8)
plt.show()
```

```python
x_values = df.year
y_values = (df.non_zero_tip_count / df.total) * 100

# Plot the values
plt.plot(x_values, y_values, marker='o', linestyle='-', color='b')
plt.title('The percent of trips where tip was given')
plt.xlabel('year')
plt.ylabel('percent of rides with tip [%]')
plt.grid(True, alpha=0.8)
#plt.savefig("./figures/cleaning/year_tip.pdf")
plt.show()
```

```python
x_values = df.year
y_values = df.non_zero_tip_avg

# Plot the values
plt.plot(x_values, y_values, marker='o', linestyle='-', color='b')
plt.title('The average amount of (non-zero) tips in each year')
plt.xlabel('year')
plt.ylabel('average tip')
plt.grid(True, alpha=0.8)
plt.show()
```

```python
x_values = df.year
y_values = df.tip_avg

# Plot the values
plt.plot(x_values, y_values, marker='o', linestyle='-', color='b')
plt.title('The average amount of tips in each year')
plt.xlabel('year')
plt.ylabel('average tip')
plt.grid(True, alpha=0.8)
plt.show()
```
